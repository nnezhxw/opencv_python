```python
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('./01_data/ml.png')
img2 = cv.imread('./01_data/opencv-logo.png')

# addWeighted
# 
img1 = img1[:230, :186]
img2 = img2[:230, :186]

# addWeighted 
dst = cv.addWeighted(img1, 0.7, img2, 0.2, 0)


# plt.figure(figsize=(8, 3))

plt.subplot(1, 3, 1)
plt.imshow(img1[:,:,::-1])
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 2)
plt.imshow(img2[:,:,::-1])
plt.xticks([])
plt.yticks([])

plt.subplot(1, 3, 3)
plt.imshow(dst[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```
