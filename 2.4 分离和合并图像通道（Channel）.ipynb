```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('./01_data/messi5.jpg')

# 分离图像通道
b, g, r = cv.split(img)

# 恐怕下面这个方法显示出来的图片不是实际上图片应该的样子。
# 因为cv.imshow()默认显示3个通道的颜色，而现在b, g, r中都只有一个通道的颜色，
# 所以可能cv.imshow()无法判断b, g, r究竟哪个是b, 哪个是g, 哪个是r
# cv.imshow('b', b)
# cv.imshow('g', g)
# cv.imshow('r', r)

# while(1):
#     if (cv.waitKey(10) & 0xFF) == 27:
#         break
        
# cv.destroyAllWindows()
```

```python
# 合并三个通道
img = cv.merge((b, g, r))

plt.imshow(img[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```
