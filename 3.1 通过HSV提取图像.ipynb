```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img = cv.imread('./01_data/opencv-logo.PNG')

# 将BGR图像转换成HSV图像
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

lower_blue = np.array([110, 50, 50])
upper_blue = np.array([130, 255, 255])

# 根据hsv中的蓝色部分生成一个mask
mask = cv.inRange(hsv, lower_blue, upper_blue)

# 提取出原图像（img）中蓝色的部分
res = cv.bitwise_and(img, img, mask = mask)

plt.subplot(1, 2, 1)
plt.imshow(img[:,:,::-1])
plt.xticks([])
plt.yticks([])

plt.subplot(1, 2, 2)
plt.imshow(res[:,:,::-1])
plt.xticks([])
plt.yticks([])
plt.show()
```
