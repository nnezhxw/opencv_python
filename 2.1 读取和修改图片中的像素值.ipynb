```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

# 读取图片
img = cv.imread('./01_data/messi5.jpg')

# 读取并输出像素点（100, 100）的值
px = img[100, 100]
print(px)
```

```python
# 读取蓝色通道（Blue Channel）上像素点的颜色值
# blue = img[100, 100, 0]
# 一个更好的读取像素点的方法
blue = img.item(100, 100, 0)
print(blue)
```

```python
# 修改像素点(100,100)的值
img[100, 100] = [0, 0, 0]
print(img[100, 100])
```
